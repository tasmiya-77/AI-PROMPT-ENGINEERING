#  AI Assistant Development (Offline)

This project is an offline, CLI-based AI Assistant simulation built using Python. It helps in:
- Answering factual questions using pre-defined Q&A
- Summarizing text using simple logic
- Generating creative content (short stories, poems)

This approach helps practice project structuring and logic building without using paid APIs.


```python
# Vault of Codes AI Assistant - Offline Version

def answer_questions():
    qa = {
        "what is the capital of japan?": "The capital of Japan is Tokyo.",
        "who is the prime minister of india?": "The Prime Minister of India is Narendra Modi.",
        "what is the largest ocean?": "The Pacific Ocean is the largest ocean on Earth."
    }
    question = input("Ask me a factual question: ").strip().lower()
    print("\nAnswer:", qa.get(question, "Sorry, I don't know the answer to that question.\n"))

def summarize_text():
    text = input("Paste the paragraph you want summarized:\n")
    sentences = text.split('. ')
    summary = '. '.join(sentences[:2]) + '.'
    print("\nSummary:", summary, "\n")

def generate_creative_content():
    print("\nChoose creative content type:")
    print("1. Short Story")
    print("2. Poem")
    choice = input("Enter your choice (1 or 2): ")
    if choice == '1':
        print("\nOnce upon a time, a small puppy got lost in the city but found its way home using its keen sense of smell.\n")
    elif choice == '2':
        print("\nRoses are red,\nViolets are blue,\nLearning each day,\nIs good for you.\n")
    else:
        print("Invalid choice.\n")

while True:
    print("\n--- Vault of Codes AI Assistant (Offline) ---")
    print("1. Answer Questions")
    print("2. Summarize Text")
    print("3. Generate Creative Content")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        answer_questions()
    elif choice == '2':
        summarize_text()
    elif choice == '3':
        generate_creative_content()
    elif choice == '4':
        print("Exiting Assistant. Goodbye!")
        break
    else:
        print("Invalid choice. Please select 1-4.\n")
```
